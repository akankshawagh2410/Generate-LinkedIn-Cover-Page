```python
from wordcloud import WordCloud
import PyPDF2
import matplotlib.pyplot as plt
```

```python
# extracting text from PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text() + ' '
    return text

#Resume pat
resume_path = "Resume_Akanksha Wagh_WA.pdf"

# Extract text
resume_text = extract_text_from_pdf(resume_path)
```

```python
# Define stopwords
stopwords = set(WordCloud().stopwords)
stopwords.update(["the", "and", "for", "with", "you", "are", "from", "this", "that"])
```

```python
# Generate word cloud
wordcloud = WordCloud(
    width=1584,
    height=396,
    background_color='white',
    colormap='Blues',
    stopwords=stopwords
).generate(resume_text)
```

```python
# Save image
output_path = "C:/Users/akank/OneDrive - stevens.edu/Desktop/Resume_WordCloud.png"
wordcloud.to_file(output_path)
```

```python
# Display image
plt.figure(figsize=(15, 4))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("LinkedIn Word Cloud Banner", fontsize=16)
plt.show()

output_path
```
